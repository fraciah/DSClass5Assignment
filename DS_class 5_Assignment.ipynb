```python
NC = ['Benue State', 'Kogi', 'Kwara', 'Nasarawa', 'Niger', 'Plateau', 'FCT']
SS = ['Akwa', 'Bayelsa', 'Cross River', 'Rivers', 'Delta', 'Edo']
SE = ['Abia', 'Anambra', 'Ebonyi', 'Enugu', 'Imo']
SW = ['Ekiti', 'Lagos', 'Ogun', 'Ondo', 'Osun', 'Oyo']
NE = ['Adamawa', 'Bauchi', 'Borno', 'Gombe', 'Taraba', 'Yobe']
NW = ['Jigawa', 'Kaduna', 'Kano', 'Katsina', 'Kebbi', 'Sokoto', 'Zamfara']
```

```python
def region(state):
    NC = ['Benue State', 'Kogi', 'Kwara', 'Nasarawa', 'Niger', 'Plateau', 'FCT']
    SS = ['Akwa', 'Bayelsa', 'Cross River', 'Rivers', 'Delta', 'Edo']
    SE = ['Abia', 'Anambra', 'Ebonyi', 'Enugu', 'Imo']
    SW = ['Ekiti', 'Lagos', 'Ogun', 'Ondo', 'Osun', 'Oyo']
    NE = ['Adamawa', 'Bauchi', 'Borno', 'Gombe', 'Taraba', 'Yobe']
    NW = ['Jigawa', 'Kaduna', 'Kano', 'Katsina', 'Kebbi', 'Sokoto', 'Zamfara']
    if state in NC:
        return 'North Central'
    elif state in SS:
        return 'South South'
    elif state in SE:
        return 'South East'
    elif state in SW:
        return 'South West'
    elif state in NE:
        return 'South South'
    else:
        return 'North West'
```

```python
data['Region']=data['STATES'].apply(region)
data
```

```python
es=data.groupby('Region')
for name,group in es:
    print(name,'\n')
    print(group,'\n')
    print('This region has ',len(group),' states.\n')
#Total number of states in each region
```
